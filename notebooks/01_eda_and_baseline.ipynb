```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sys
import os

# path to internal tools
sys.path.append(os.path.abspath('../src'))
from features import create_physics_features

# load Data
df_raw = pd.read_csv('../data/raw.csv')

# engineering Layer
df_engineered = create_physics_features(df_raw)

# correlation check
plt.figure(figsize=(10, 6))
viz_cols = ['Energy_Consumption_kWh', 'Speed_ms', 'Drag_Proxy', 'Incline_Work_Proxy', 'Thermal_Stress_Index']
sns.heatmap(df_engineered[viz_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation: Physics Features vs. Energy Consumption")
plt.show()

# export
df_engineered.to_csv('../data/processed_engineered.csv', index=False)
```
